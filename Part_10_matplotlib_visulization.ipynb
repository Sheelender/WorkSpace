### Package name: Matplotlib(matplotlibrary)

```python
import matplotlib.pyplot as plt
import numpy
```

- plt.plot():Create a plot object
- plt.title():And a title for plot object
- plt.xlable():specify lables of x-axis
- plt.yaxis():specify lables of y-axis
- plt.show():Display the plot

# line plot

```python
# suppose you want to plot a sales plot
# sales happens in every month

sales=[100,122,203,304,405,506,96,700,600,88,900,1000]
len(sales)
import numpy as np
months=np.arange(1,13)
months
```

```python
plt.plot(months,sales,marker='*')
plt.xlabel('month')
plt.ylabel('sales')
plt.title("sales view")
plt.savefig('Months_sales.jpg')

plt.show() # always in the end
```

```python
plt.plot(months,sales,marker='*')
plt.xlabel('month')
plt.ylabel('sales')
plt.title("sales view")
plt.savefig('Months_sales.jpg')
plt.grid()
plt.show() # always in the end
```

```python
plt.plot(months,sales,marker='*')
plt.xlabel('month')
plt.ylabel('sales')
plt.title("sales view")
plt.savefig('Months_sales.jpg')
plt.grid(linestyle='-',linewidth=1,color='green')
plt.show() # always in the end
```

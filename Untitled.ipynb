$CASE-1$:
    
**Observatios**

- To fit a linear regression model, we select those features which have high correlation with our target variable MEDV

- By seeing the correlation matrix RM has highly postively correlated i.e. 0.70

- LSTAT highly negatively correlated i.e. -0.74

```python
plt.subplot(1,2,1)
plt.scatter(df['LSTAT'],df['MEDV'])
plt.xlabel("MEDV")
plt.ylabel('LSTAT')

plt.subplot(1,2,2)
plt.scatter(df['RM'],df['MEDV'])
plt.xlabel("MEDV")
plt.ylabel('RM')
```

**step:1**: Divide data into input and output

```python
#  preparing data
X=df[['LSTAT','RM']]  # input data having inly two columns i.e RM and LSTAT
Y=df['MEDV']          # Output data having traget column i.e MEDV

```

**Step:2**:Divide data into train and test data

```python
# Will divide data into train data and test data
# train data===> X_train and y_train
# test data====> x_test and y_test
# 
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=1234)
```

```python
X_test
```

```python
X_train
```

```python
X_train.shape,X_test.shape
```

**step:3** Model creation

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error


# step:import the package
# step:save the package
# apply fit transform


linear_model=LinearRegression()
# fit the model / model creation based on train data
linear_model.fit(X_train,Y_train)
```

```python
dir(linear_model)
```

#MEDV=b0+b1*LSTAT+b2*RM

```python
linear_model.coef_
```

```python
linear_model.intercept_
```

MEDV=3.795751947754855

**Step:4** -Prediction on test data

```python
# Get model prediction by providing X_test data
# This will give some prediction i.e.y_test_prediction
# will compare these prediction with original Y_test data
Y_test_prediction=linear_model.predict(X_test)
Y_test_prediction
```

```python
print(Y_test_prediction[:10])
print(Y_test[:10].values)
```

**Step:5** Model evaluation

```python
mse=mean_squared_error(Y_test,Y_test_prediction)
rmse=np.sqrt(mse)
r_square=r2_score(Y_test,Y_test_prediction)
print("MSE",mse)
print("rmse",rmse)
print("r_square",r_square)
```
